```python


import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into a DataFrame
data = pd.read_csv('C:/Users/maxru/OneDrive/Wrestling Mindset/Wrestling Mindset 1on1 Program Subscribers 2022-2024 - Sheet1 (Trial).csv')

# Filter out rows with NaT (missing values) in Date2024
data_2024_valid = data.dropna(subset=['Date2024'])

# Extracting data for each set of axes
date_2022 = pd.to_datetime(data['Date2022'] + '/2022')
subscribers_2022 = data['Subscribers2022']

date_2023 = pd.to_datetime(data['Date2023'] + '/2022')
subscribers_2023 = data['Subscribers2023']

date_2024 = pd.to_datetime(data_2024_valid['Date2024'] + '/2022')
subscribers_2024 = data_2024_valid['Subscribers2024']

# Plotting
plt.figure(figsize=(18, 8))

# Plotting first set of data
plt.plot(date_2022, subscribers_2022, 'o-', label='2022')

# Plotting second set of data
plt.plot(date_2023, subscribers_2023, 'o-', label='2023')

# Plotting third set of data
plt.plot(date_2024, subscribers_2024, 'o-', label='2024')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Subscribers')
plt.title('Subscribers Over Time')
plt.legend()

# Annotating each data point with its value
for x, y in zip(date_2022, subscribers_2022):
    plt.text(x, y, str(y), ha='left', va='bottom', rotation=45)

for x, y in zip(date_2023, subscribers_2023):
    plt.text(x, y, str(y), ha='left', va='bottom', rotation=45)

for x, y in zip(date_2024, subscribers_2024):
    plt.text(x, y, str(y), ha='left', va='bottom', rotation=45)

# Setting x-axis limits to cover from January 1st to December 31st
plt.xlim(pd.to_datetime('2022-01-01'), pd.to_datetime('2022-12-31'))

# Formatting x-axis ticks
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%m/%d'))

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()










```
